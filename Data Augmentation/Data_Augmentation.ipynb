## Data Augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
# Hyperparameters
datagen = ImageDataGenerator(rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             fill_mode='nearest')
```

```python
# Uploading image
img = load_img('/content/drive/MyDrive/dog-breed-identification/dogs/YellowLabradorLooking.jpg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)
x.shape
```

```python
# Generate 50 different image from uploaded image
i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='drive/MyDrive/dog-breed-identification/dogs',
                          save_format='jpg'):
  i += 1
  if i = 50:
    break
```

## Data Augmentation With CIFAR10 Dataset 

```python
# Import libraries
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from keras import backend as K
import numpy as np
```

```python
# Download dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train/=255
X_test/=255
```

```python
# Display data
datagen = ImageDataGenerator()
datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=112):
  for i in range(0,4):
    plt.subplot(220 +1 +i)
    plt.imshow(X_batch[i])
  plt.show()
  break
```

### Rotating

```python
datagen = ImageDataGenerator(rotation_range = 136)
datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=112):
  for i in range(0,4):
    plt.subplot(220 +1 +i)
    plt.imshow(X_batch[i])
  plt.show()
  break
```

### Shifting Vertically

```python
datagen = ImageDataGenerator(height_shift_range=0.5)
datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=112):
  for i in range(0,4):
    plt.subplot(220 +1 +i)
    plt.imshow(X_batch[i])
  plt.show()
  break
```

### Horizontal Symmetric

```python
datagen = ImageDataGenerator(horizontal_flip=True)
datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=4, seed=112):
  for i in range(0,4):
    plt.subplot(220 +1 +i)
    plt.imshow(X_batch[i])
  plt.show()
  break
```
